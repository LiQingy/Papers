```python
%pylab notebook
import numpy as np 
import h5py
import matplotlib.pyplot as plt
import random
import sys
sys.path.append('/home/qyli/oPDF2')
```

```python
fileCM = '../proxy/cluster/G3X-CM-masscomplete-ALL.txt'
filefit = '../proxy/cluster/GXsub_DM_fitmc_rcin200_100000_TMP.txt'
dataCM = np.loadtxt(fileCM)
datafit = np.loadtxt(filefit)
import oPDFplus as opd
mmtrue = datafit[:,0] / dataCM[:,2] * 1e15
cctrue = datafit[:,1] / dataCM[:,6]
loc = opd.sigma3sel(np.log10(mmtrue),np.log10(cctrue))
```

```python
maind = np.loadtxt('../proxy/GadgetX_DS_main')[loc,1:]
```

```python
def calall_cov(nphy):

	phyall = [maind[:,0],maind[:,1],maind[:,2],maind[:,3],maind[:,5],maind[:,6],maind[:,8],maind[:,10],maind[:,16],maind[:,24]]
	label = [r'$\eta$',r'$\Delta_r$',r'$f_{\rm s}$',r'$\chi_{\rm DS}$',r'$\dot{M}_\mathrm{a}$',r'$f_{\rm ste}$','$M1$','$r1$',r'$\sigma_V$','$T_{500}$']
	covall = np.zeros(shape = (nphy,nphy))

	for i in range(nphy):
		for j in range(nphy-1-i):
			j = j+i+1
			pro1 = phyall[i]
			pro2 = phyall[j]
			# if label[i] == 'Ma' or label[j] == 'Ma':
			# 	pro1 = pro1[locMa0]
			# 	pro2 = pro2[locMa0]
			cov = np.cov(pro1,pro2)
			coveff = cov[0][1] / cov[0][0]**0.5 / cov[1][1]**0.5
			print('The correlation coefficient between %s and %s is %s' %(label[i],label[j],coveff))

			covall[i,j] = coveff
		covall[i,i] = 1
	covall = np.around(covall, 2)
	print(covall)
	return covall, label
```

```python
def plot_d3(ax):
    nphy = 10
    covall,label = calall_cov(nphy)

    covmap = ax.imshow(np.abs(covall), cmap = 'Blues',vmin = 0, vmax = 1)
#     cbar = ax.colorbar()
#     cbar.ax.set_ylabel(r'$|\gamma|$', fontsize = 12)

    ax.set_xticklabels(label,fontsize = 14)
    ax.set_yticklabels(label,fontsize = 14)
    ax.set_xticks(np.arange(len(label)))
    ax.set_yticks(np.arange(len(label)))

    for i in range(len(label)):
        for j in range(nphy-i):
            j = nphy - 1 - j 
            text = ax.text(j, i, '%s' %covall[i, j],
                            ha="center", va="center", color="k")				
    # plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
    #         rotation_mode="anchor")
    ax.invert_xaxis()
    ax.invert_yaxis()

    ax.tick_params( labelsize = 11, direction = 'in')
    bwith = 1.2 #边框宽度设置为2
    ax.spines['bottom'].set_linewidth(bwith)
    ax.spines['left'].set_linewidth(bwith)
    ax.spines['top'].set_linewidth(bwith)
    ax.spines['right'].set_linewidth(bwith)
```

## plot observation indicators and D3 indicators

```python
def calob_cov():
    dSZ = np.loadtxt('../proxy/GadgetX_DS_SZ')[loc,1:]
    dXray = np.loadtxt('../proxy/GadgetX_DS_Xray')[loc,1:]
    dOffset = np.loadtxt('../proxy/GadgetX_DS_Offset')[loc,1:]

    phyall = [dXray[:,2],dXray[:,-1],dSZ[:,-1],dOffset[:,1],dOffset[:,5],dOffset[:,2],dOffset[:,6]]
#     phyall = [dSZ[:,0],dSZ[:,1],dSZ[:,2],dSZ[:,3],dSZ[:,4],dSZ[:,5],dSZ[:,6]]
    d3fea = [maind[:,0], maind[:,5], maind[:,9], maind[:,-3]]
#     labelall = ['1','2','3','4','5','6','7']
    labelall = [r'$P_{\rm Xray}$', r'$M_{\rm Xray}$', r'$M_{\rm SZ}$', r'$R_{\rm SZ}$', r'$O_{\rm BCG-SZ}$', r'$R_{\rm X}$',r'$O_{\rm BCG-Py}$']
    labeld3 = [r'$\eta$',r'$\dot{M}_\mathrm{a}$','$M2$','$T_{500}$']
    covall = np.zeros(shape = (4,7))

    for i in range(4):
        for j in range(7):
            pro1 = d3fea[i]
            pro2 = phyall[j]
            cov = np.cov(pro1,pro2)
            coveff = cov[0][1] / cov[0][0]**0.5 / cov[1][1]**0.5
#             print('The correlation coefficient between %s and %s is %s' %(labeld3[i],labelall[j],coveff))

            covall[i,j] = coveff
    covall = np.around(covall, 2)
    print(covall)
    return covall, labelall, labeld3
```

```python
def plot_obcov(ax):
    covall,labelall, labeld3 = calob_cov()

    covmap = ax.imshow(np.abs(covall), cmap = 'Blues',vmin = 0, vmax = 1)
#     cbar = ax.colorbar()
#     cbar.ax.set_ylabel(r'$|\gamma|$', fontsize = 12)

    ax.set_xticklabels(labelall,fontsize = 12)
    ax.set_yticklabels(labeld3,fontsize = 10)
    ax.set_xticks(np.arange(len(labelall)))
    ax.set_yticks(np.arange(len(labeld3)))

    for i in range(len(labeld3)):
        for j in range(len(labelall)):
            text = ax.text(j, i, '%.2f' %covall[i, j],
                            ha="center", va="center", color="k")
	# plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
 #         rotation_mode="anchor")
# 	ax.invert_xaxis()
# 	ax.invert_yaxis()

    ax.tick_params( labelsize = 11, direction = 'in')
    bwith = 1.2 #边框宽度设置为2
    ax.spines['bottom'].set_linewidth(bwith)
    ax.spines['left'].set_linewidth(bwith)
    ax.spines['top'].set_linewidth(bwith)
    ax.spines['right'].set_linewidth(bwith)
    
    return covmap
```

## plot a combination

```python
plt.figure(figsize = (10,4.5))
axd3 = plt.subplot2grid((4,13),(0,0),colspan = 6,rowspan = 4)
axdob = plt.subplot2grid((4,13),(0,6),colspan = 7,rowspan = 4)
# axcolorbar = plt.subplot2grid((4,10),(0,),colspan = 5,rowspan = 4)

plot_d3(axd3)
covmap = plot_obcov(axdob)
plt.colorbar(covmap)

plt.tight_layout()
plt.show()
plt.savefig('./figs/cc.pdf')
```
